```python
#we are going to remove HDI for year as it has null, and country-year as it is repeated
case.drop(['HDI for year','country-year'],axis=1,inplace=True)
case
```

```python
#new data frame with the columns that was asked for 
newcase=case[['country','sex','age','suicides/100k pop']]
newcase
```

```python
#renaming the column as suicide rates
newcase=newcase.rename(columns = {"suicides/100k pop" :"suicide_rates"})
newcase
```

```python
#removing the rows whose dataset has 0 suicide rates
newcase.drop(newcase.loc[newcase['suicide_rates']==0.00].index, inplace=True)
```
